##**3. Cleaning duration**
Currently in format hh:mm:ss, we want to convert this into a standard format (Seconds) which can be incorporated better for our analysis

```python
duration = list((df['duration'].values))
type(duration)
```

```python
Std_duration=[]
for i in df['duration']:
  if len(i.split(':'))==2:
    mins = int(i.split(':')[0])
    secs = int(i.split(':')[1])
    Std_duration.append(mins*60+secs)
  else:
    hrs = int(i.split(':')[0])
    mins = int(i.split(':')[1])
    secs = int(i.split(':')[2])
    Std_duration.append(hrs*60*60+mins*60+secs)

print(type(Std_duration)) #Checking the type of our new duration
print(len(Std_duration)) #Checking the number of rows in our new duration

df['Std_duration']=Std_duration #Adding this list as a new column in our dataframe
```

```python
#Checking if new column 'Std_duration has been added to our database

df.head()
```

##**4. Cleaning Singers/ Singer_id**

*   Removing the pipe operator ('|') and the repetitive string '/artist/'
*   Inserting a comma between the singers in case of duets



```python
df['singer'][0]
```

```python
df['singer_id'] = [i.replace('/artist/','').replace('|',',') for i in df['singer_id']]
df['singer'] = [i.replace('|',',') for i in df['singer']]
df.head()
```

**5.17 : Finding unique number of singers**

```python
#We will first find out the list of unique singers only

singers_c = []

for i in df['singer_id']:
  singers_c += i.split(',')

singers_c = list(set(singers_c))

print(len(singers_c))
```

**5.18 : Finding number of songs sung by each singer**

```python
lst = []

for i in tqdm(singers_c):
  f = 0
  for j in df['singer_id']:
    if i in j:
      f += 1

  lst.append([i,f])

lst
```

**5.19 : Singers with maximum number of songs (by restructuring the dataset)**

```python
songs = pandas.DataFrame(lst, columns = ['singer_id','songs'])
songs.head()
songs.sort_values(by = 'songs', ascending = False).head(10)
```
